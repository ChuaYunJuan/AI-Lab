```python
#Question 1: Create Panda Series 
import pandas as pd
step_data = [3620, 7891, 9761, 3907, 4338, 5373]  # create a Python list
step_counts = pd.Series (step_data, name = 'steps') # converts the the list "step_data" into a pandas "Series"
# Series = 1D array object that can hold any data type  within the same structure
### pd.Series (data, name = 'series_name') 
# 'pd.Series' = constructor to create "Series"
# name = 'series_name' ; to assign a name to the Series 

print(step_counts)

# Extra: 
# The index for the 'Series' is generated automatically and starts at 0. So indices will be '[0,1,2,3,4,5]'
```

```python
#Question 2: Add Data Range to Series 
import pandas as pd
step_data = [3620, 7891, 9761, 3907, 4338, 5373, 7721]  
step_counts = pd.Series (step_data, name = 'steps')

step_counts.index = pd.date_range('20240501',periods = 7)
### step_counts.index
# The index of step_counts [0,1,2,3,4,..] being replaced to date range

### pd.date_range('YYYYMMDD', periods = )
# 'pd.date_range' = function of pandas library that generate a sequences of dates
# '20240501' = start date for date range in format of YYYYMMDD
# ' periods = ' = number of dates to be generated

print(step_counts)
```

```python
#Question 3: Select Data by Index Value
import pandas as pd
step_data = [3620, 7891, 9761, 3907, 4338, 5373, 7721]  
step_counts = pd.Series (step_data, name = 'steps')
step_counts.index = pd.date_range('20240501',periods = 7)

# Method 1: Select like a dictionary
print(step_counts['2024-05-07'])

# Method 2: Select by an index position -- like an array
print(step_counts.iloc[6])
### '.iloc' = to access the value by its position in Series (use like an array)

# Method 3: Select all date in May
print(step_counts['2024-05'])
```

```python
# Question 4: View and Convert Data Type
import pandas as pd
step_data = [3620, 7891, 9761, 3907, 4338, 5373, 7721]  
step_counts = pd.Series (step_data, name = 'steps')
step_counts.index = pd.date_range('20240501',periods = 7)

print(step_counts.dtypes)
### '.dtypes' = View data type

step_counts = step_counts.astype(float) 
### '.astype()' = change the original data type to a new one
print(step_counts)
```

```python
#Question 5: Create Pandas Dataframe

import pandas as pd
# Array 1 
step_data = [3620, 7891, 9761, 3907, 4338, 5373]  
step_counts = pd.Series (step_data, name = 'steps')

# Array 2
cycling_data = [10.7, 0, None, 2.4, 15.3, 10.9, 0, None]

# Joined 2 Array of Data
joined_data = list(zip(step_data, cycling_data))
### 'zip()' = takes 2 or more iterables
### 'list()' = used to converts the resulting iterator into a list of tuples

# Create Dataframe
activity_df = pd.DataFrame(joined_data)
### 'pd.DataFrame()' = used to create a pandas DataFrame from the list of tuples in ()
print(activity_df)
```

```python
# Question 6: Add Labeled Column and Index 
import pandas as pd
step_data = [3620, 7891, 9761, 3907, 4338, 5373]  
step_counts = pd.Series (step_data, name = 'steps')
cycling_data = [10.7, 0, None, 2.4, 15.3, 10.9, 0, None]

joined_data = list(zip(step_data, cycling_data))

activity_df = pd.DataFrame(joined_data,
            index = pd.date_range('20240501', periods = 6), 
            columns = ['Walking', 'Cycling'] )  # name the columns of DataFrame

# pd.DataFrame(tuples_list, 
#             index = 
#             column = 
#             )

print(activity_df)
```

```python
# Question 7: Select Data by Index Value 
import pandas as pd
step_data = [3620, 7891, 9761, 3907, 4338, 5373]  
step_counts = pd.Series (step_data, name = 'steps')
cycling_data = [10.7, 0, None, 2.4, 15.3, 10.9, 0, None]

joined_data = list(zip(step_data, cycling_data))

activity_df = pd.DataFrame(joined_data, index = pd.date_range('20240501', periods = 6),
                           columns = ['Walking', 'Cycling'] ) 

# Method 1: Select row of data by index name 
print(activity_df.loc['2024-05-04'])
### '.loc' = used to access rows or column by labels or boolean array

# Method 2: Select row of data by an index position -- like an array
print(activity_df.iloc[4])
### '.iloc' = used to access rows or column by integer position

# Method 3: Select column of data by name OR integer 
##print(activity_df['Walking']) 
##print(activity_df.Walking)
print(activity_df.iloc[:,0])

### ' : ' = Select all rows 
### ' 0 ' = Select first column
```

```python
# Question 9: Create New Column
import pandas as pd
filepath = 'Iris_Data.csv'

data = pd.read_csv(filepath)

data['sepal_area'] = data.sepal_length * data.sepal_width
# create a new column named 'sepal_area'

print(data.iloc[:5, -3:])
# ' :5 ' = specify first 5 rows 
# ' -3: ' = specify last 3 column
```

```python
# Question 10: Apply Function to Column
import pandas as pd 
filepath = 'Iris_Data.csv'

data = pd.read_csv(filepath)

data['abbrev'] = (data.species.apply(lambda x: x.replace('Iris-', ' ')))
# create a new column named 'abbrev'
### 'data.species' = get access to the 'species' column in DataFrame
### 'apply()' = function to apply a function each element in the column
### ' lambda x: x.replace(' ', ' ') ' = lambda function take each element 'x' from the 'species' and applies the 'replace()' 
print(data.iloc[:5,-3:])

# Join the first 2 rows and last 2 rows
small_data = pd.concat([data.iloc[:2], data.iloc[-2:]])
# 'data.iloc[:2]' = first 2 rows of DataFrame
# 'data.iloc[-2:]' = last 2 rows of DataFrame
# 'pd.concat()' = concatenate 2 DataFrame slices along the default axis (row)

print(data.iloc[:, -3:])
```

```python
# Question 11: Aggregated Statistics
import pandas as pd 
filepath = 'Iris_Data.csv'

data = pd.read_csv(filepath)

# In order to count in DataFrame, we can use the size method 
group_sizes = (data.groupby('species').size())
# 'data.groupby(' unknown ')' = groups the DataFrame by the values in the 'unknown' column
# '.size ()' = method that calculates the number of elements (rows) in each group created by 'groupby'

print(group_sizes)
```

```python
# Question 12: Statistical Calculations 
import pandas as pd 
filepath = 'Iris_Data.csv'

data = pd.read_csv(filepath)

# To calculate mean in a DataFrame
print(data.iloc[:,:4].mean())
# '.mean()'

# To calculate median on a Series 
print(data.petal_length.median())
# '.medioan()'

# To calculate mode on a Series
print(data.petal_length.mode())

# To calculate standard deviation, variance, standard error of mean (SEM)
print (data.petal_length.std())
print (data.petal_length.var())
print (data.petal_length.sem())

# To calculate quantile = Value that split sorted data (3-Quantile split data into 4 groups)
print(data.iloc[:,:4].quantile(0))

# In orde to present multiple calculations 
print(data.iloc[:, :4].describe())
```

```python
# Question 13: Matplotlib - To plot single layer of Data
import matplotlib.pyplot as plt

plt.plot(data.sepal_length, data.sepal_width, ls = '', marker = 'o')
#'data.sepal_length' = data for x axis
#'data.sepal_width' = data for y axis
# 'ls='' ' = line style will be an empty string (no line will be drawn to connect the points)
# ' marker = 'o''  = each data point should be represented as a circle

# Set x & y label, and title 
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')

# Display the plot 
plt.show()
```

```python
# Question 14: Matplotlib - To plot multiple layer of Data

import matplotlib.pyplot as plt

plt.plot(data.sepal_length, data.sepal_width, ls = '', marker = 'o', label = 'sepal')
plt.plot(data.petal_length, data.petal_width, ls = '', marker = 'o', label = 'petal')

# Set x & y label, and title 
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Display the plot with legend 
plt.legend()
plt.show()
```

```python
# Question 15: Matplotlib - To plot histogram
import matplotlib.pyplot as plt

plt.hist (data.sepal_length, bins = 25)
# bins = 25 -- number of bins (intervals) into which the data wull be divided
# '25' --- the range of data will be divided into 25 equal-width bins

# Add Labels
plt.xlabel ('Sepal Length')
plt.ylabel ('Number of Irises')

# Display the plot 
plt.show()
```
